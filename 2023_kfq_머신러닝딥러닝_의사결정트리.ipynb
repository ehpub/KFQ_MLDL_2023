```python
train_df = train_df.drop('district',axis=1)
test_df = test_df.drop('dong',axis=1)
```

```python
x_train = train_df.drop('label',axis=1)
y_train = train_df['label']
```

```python
x_test = test_df.drop('label',axis=1)
y_test = test_df['label']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
le.classes_
```

```python
def display_decision_surface(model,x,y):
  x_min = x.longitude.min() - 0.0
  x_max = x.longitude.max() + 0.01
  y_min = x.latitude.min() - 0.01
  y_max = x.latitude.max() + 0.01

  n_classes = len(le.classes_)
  plot_colors = "rywb"
  plot_step = 0.001
  xx,yy = np.meshgrid(np.arange(x_min,x_max,plot_step),
                    np.arange(y_min,y_max,plot_step))
  g_test = np.c_[xx.ravel(),yy.ravel()]
  pred = model.predict(g_test)
  pred = pred.reshape(xx.shape)
  plt.contourf(xx,yy,pred)

  for i, color in zip(range(n_classes),plot_colors):
    idx = np.where(y==i)
    plt.scatter(x.loc[idx].longitude,
              x.loc[idx].latitude,
              c = color,
              label = le.classes_[i],
              edgecolor='black',
              s=200
              )
  plt.show()
```

```python
display_decision_surface(model,x_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,20))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred = model.predict(x_test)
accuracy_score(y_test,pred)
```
